```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
plt.style.use('ggplot')
plt.style.use('seaborn-bright')
```

```python
loc_info = pd.read_excel("data/8.1 MSAN location data.xls")  
loc_info
```

```python
lte_loc_info = pd.read_excel("data/8.2 LTE location data.xlsx")  
lte_loc_info
```

```python
loc_info['Location_Identifier'] = loc_info['MSAN'].astype(str)+ "_" + loc_info['LOCATION'].astype(str) + "_" +loc_info['EQUP_ID'].astype(str) +"_" +loc_info['EQ_INDEX'].astype(str)
loc_info.drop(["LOCATION","EQUP_ID","EQ_INDEX","MSAN"], axis = 1, inplace = True)
```

```python
loc_info = loc_info.drop_duplicates(["CR.hash","DISPLAYNAME.hash"], keep="last")
```

```python
lte_loc_info = lte_loc_info[["Number.hash","NODE NAME"]]
lte_loc_info.rename(columns={"Number.hash":"event_source.hash","NODE NAME":"Location_Node"},inplace=True)
lte_loc_info.drop_duplicates(["event_source.hash","Location_Node"],inplace=True)
lte_loc_info
```

### Grouping by locations

```python
grouped_loc_df = loc_info[["CR.hash","Location_Identifier" ]].groupby('CR.hash', as_index=False).agg(list)
```

```python
grouped_loc_df["Location_Identifier"] = grouped_loc_df["Location_Identifier"].apply(lambda x:  list(set(x)))
```

```python
grouped_loc_df
```

```python
grouped_lte_loc_df = lte_loc_info.groupby('event_source.hash', as_index=False).agg(list)
grouped_lte_loc_df["Location_Node"] = grouped_lte_loc_df["Location_Node"].apply(lambda x:  list(set(x)))
grouped_lte_loc_df
```

### This shows no harm if we drop by unique User Ids

```python
lte_loc_info.drop_duplicates("event_source.hash",inplace=True)
lte_loc_info
```

```python
loc_info.drop_duplicates("CR.hash",inplace=True)
loc_info
```

```python
loc_info.to_csv("data/user_profile/User_Locations.csv")
```

```python
lte_loc_info.to_csv("data/user_profile/User_LTE_Locations.csv")
```

### Grouping by No of Users

```python
loc_info["User_Count"] = 1
gropued_by_locations = loc_info.groupby(['Location_Identifier'], axis=0,as_index=False).sum()
gropued_by_locations = gropued_by_locations[['Location_Identifier','User_Count']]
gropued_by_locations
```

```python
gropued_by_locations.to_csv("data/user_profile/Users_Grouped_By_Locations.csv")
```

```python
plot_sorted = gropued_by_locations.sort_values("User_Count",ascending=False).head(30)
plot_sorted.plot(kind="bar", figsize=(18,10))
```

```python
lte_loc_info["User_Count"] = 1
lte_gropued_by_locations = lte_loc_info.groupby(['Location_Node'], axis=0,as_index=False).sum()
lte_gropued_by_locations = lte_gropued_by_locations[['Location_Node','User_Count']]
lte_gropued_by_locations
```

```python
lte_gropued_by_locations.to_csv("data/user_profile/Users_Grouped_By_LTE_Locations.csv")
```

```python
plot_sorted = lte_gropued_by_locations.sort_values("User_Count",ascending=False).head(30)
plot_sorted.plot(kind="bar", figsize=(18,10))
```
