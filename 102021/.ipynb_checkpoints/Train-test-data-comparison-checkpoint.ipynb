```python
import pandas as pd
from math import nan
```

```python
# test accuracy ratio
def rec_pack(ori_pack, pack_list):
    ori_pack = ori_pack.replace("_"," ")
    for pack in pack_list:
        pack = pack.replace("_"," ")
        if ori_pack in pack:
            return "YES"
    
def rec_pack_rank(ori_pack, pack_list):
    weight = 1
    ori_pack = ori_pack.replace("_"," ")
    for pack in pack_list:
        pack = pack.replace("_"," ")
        if ori_pack in pack:
            return weight
        weight = weight - (weight/5)
    return weight
```

## Peo TV

```python
result_peo_df = pd.read_csv("data/cross-selling-scored/peotv_scored_filterd.csv")
```

```python
test_peo_df = pd.read_parquet("data/cross-selling-scored/peotv_test.parquet")
```

```python
test_peo_df = test_peo_df.merge(result_peo_df, how="left", left_on="ACCOUNT_NUM.hash", right_on = "User")
```

```python
test_peo_df.dropna(subset=["User"], inplace=True)
test_peo_df.drop("User", axis=1, inplace = True)
```

```python
test_peo_df["Actual_Item_Is_Recommended"] = test_peo_df.apply(lambda x: rec_pack(x["package"], [x["Recommended Item 1"], x["Recommended Item 2"], x["Recommended Item 3"], x["Recommended Item 4"], x["Recommended Item 5"]]), axis=1)
```

```python
test_peo_df["Recommendation_Acc"] = test_peo_df.apply(lambda x: rec_pack_rank(x["package"], [x["Recommended Item 1"], x["Recommended Item 2"], x["Recommended Item 3"], x["Recommended Item 4"], x["Recommended Item 5"]]), axis=1)
```

```python
test_peo_df
```

```python
test_peo_df[(test_peo_df["Actual_Item_Is_Recommended"]== "YES")]["Actual_Item_Is_Recommended"].count()/ test_peo_df["ACCOUNT_NUM.hash"].count()
```

```python
test_peo_df["Recommendation_Acc"].value_counts()
```

```python
test_peo_df.to_csv("data/cross-selling-scored/peotv_score_validation.csv")
```
