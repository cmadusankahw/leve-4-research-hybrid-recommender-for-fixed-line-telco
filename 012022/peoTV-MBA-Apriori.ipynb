```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# import dataset
user_profile = pd.read_csv("data/user_profile/PEO_TV_Usage_Categorical_Non_Unique.csv")
user_profile
```

```python
data=user_profile.iloc[:,[1,2,12]]
data
```

```python
data["Peo_TV_Package"].value_counts()
```

## Interaction Matrix (bucket)

```python
bkt = data.groupby(["Phone_Number.hash","Peo_TV_Package"])["Usage_Rating_Scaled"].sum().unstack()
```

```python
bkt.info()
```

```python
bkt = bkt.fillna(0)
```

```python
bkt
```

```python
bkt["PEO_SILVER"].value_counts()
```

```python
def encode_units(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
# applymap -> apply a function to all values in a dataframe
bkt = bkt.applymap(encode_units)
```

```python
bkt
```

```python
bkt["PEO_SILVER"].value_counts()
```

```python
# minimum support to include -> here 0.07
frq_items= apriori(bkt,min_support=0.0001,use_colnames=True)
```

```python
frq_items
```

```python
rules=association_rules(frq_items,metric="lift",min_threshold=0.0001)
```
